````python
class Ham():
    def __init__(self,name,price):
        self.name = name
        self.price = price
class Ham_B(Ham):
    def __init__(self,name,price,kcal):
        super().__init__(name,price)
        self.kcal = kcal
b=Ham_B("버거",1000,500)
print(b.name)
print(b.price)
print(b.kcal)
````

````python
class Ham():
    def __init__(self):
        self.name = "A버거"
        self.price = 1000
class Ham_B(Ham):
    def __init__(self,kcal):
        super().__init__()
        self.name = "B버거"
        self.kcal = kcal
b=Ham_B(500)
print(b.name)
print(b.price)
print(b.kcal)
````
