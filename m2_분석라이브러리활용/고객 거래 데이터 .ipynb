고객별 속성 데이터
- cusno : 고객번호 1000개 랜덤
- gender : 0,1 정수 난수 생성
- age : 0 ~ 80사이 정수 난수 생성

거래 데이터

- product_20 : 고가제품(10), 중가제품(20), 저가제품(30)로 구분제품코드 10,20,30

- quantity_21 : 1 ~ 100 사이 정수 난수 생성
- amount_20 : price_20 * quantity_20
- amount_21 : price_21 * quantity_21
- sales : 20년 대비 21년구매금액이 증가면 1, 감소면 0

#### Q2. df를 수정하여 5가지 이상의 인사이트를 포함한 데이터 셋으로 변환하세요

#### Q3. df 를 탐색적 분석을 통하여 인사이트를 도출하세요.

- 거래일자, 요일, 시간을 10개 추가해서 총 만개의 관측치를 갖는 데이터셋 만드세요.

```python
#Q1. 고객 속성(id, 성별, 연령, 거주지역) 데이터와 거래 데이터를 통합한 데이터 프레임 df를 생성하세요.
c_id=pd.Series(range(1,1001))
li=list(range(1,1001))
data_id=random.sample(li,1000)
cusno=pd.Series(data_id)
gender=pd.Series(np.random.choice(['male','female'],1000))
age=pd.Series(np.random.randint(1,80,1000))
region=pd.Series(np.random.randint(1,10,1000))

product=pd.Series(np.random.choice([10,20,30],10000))
price=pd.Series(np.random.uniform(1000,50000,10000))
quantity=pd.Series(np.random.randint(1,100,10000))

amount = price * quantity
   
cusdf=pd.concat([c_id,cusno,gender,age,region],axis=1)
cuspu=pd.concat([product,price,quantity],axis=1)
#sales : 20년 대비 21년구매금액이 증가면 1, 감소면 0
#cusdf['sales'] = np.where(amount_20<=amount_21,1,0)
cusdf.rename(columns={0:'id',1:'cusno',2:'gender',3:'age',4:'region'},inplace=True)
cuspu.rename(columns={0:'product',1:'price',2:'quantity'},inplace=True)
cusdf.head()
```

```python
# 고객별 10개씩 행 추가
cusdf1=cusdf.copy()
for _ in range(9):
    cusdf = cusdf.append(cusdf1.loc[:],ignore_index=True)
cusdf.value_counts('cusno') #10개가 맞는지 확인
```

```python
# 거래일자,요일, 시간을 추가
# 1.data
data_list=list(pd.date_range('2020-01-01','2021-12-31',freq='D'))
cusdf['date']=np.random.choice(data_list,size=10000)

# 2. day(요일)
cusdf['day']=cusdf['date'].apply(lambda x: x.weekday())  # date별 요일을 숫자로
cusdf['day']=cusdf['day'].replace(list(range(7)),['mon','tue','wen','thu','fri','sat','sun'])

#3. month(달) 추가
cusdf['month']=cusdf['date'].dt.month

# 4. 시간
cusdf['time']=np.random.randint(1,25,size=10000)

cusdf.head()
```

```python
# Q2. df를 수정하여 5가지 이상의 인사이트를 포함한 데이터 셋으로 변환하세요
# age를 범주화 해서 age_class 컬럼을 새로 추가
def get_age(x):
    if x<20:
        x='minority'
    elif x<40:
        x='20to30year'
    elif x<60:
        x='40to50year'
    else :
        x='over60age'
    return x
cusdf['age_class'] = cusdf.age.apply(get_age)
cusdf.head()
```

```python
# 불필요한 컬럼 'id, age' 삭제
cusdf=cusdf.drop(['id','age'],axis=1)
```

```python
# 고객데이터와 구매데이터 합치기a
cusdf2=pd.concat([cusdf,cuspu],axis=1)
cusdf2.head()
```

```python
# 구매감소 인사이트 (계절 )추가하기
cusdf2.month=cusdf.month.astype('int')
def div_season(x):
    if x>=3 and x<6:
        x='spring'
    elif x>= 6 and x<9:
        x='summer'
    elif x>= 9 and x<12:
        x='fall'
    else :
        x='winter'
    return x

cusdf2['season']=cusdf2.month.apply(div_season)
cusdf2.head()
```

```python
# 인사이트를 추가 하기위해 소비 증가 사용자 함수 'spend' 만들기
def spend(x):
    return x * 1.2
# 인사이트를 추가 하기위해 소비 감소 사용자 함수 'dspend' 만들기
def dspend(x):
    return x * 0.7
# 연령,성별로 선택하여 소비증가 함수 적용하기
cusdf2.loc[cusdf2['gender']=='female','price'] = cusdf2.loc[cusdf2['gender']=='female','price'].apply(spend)
cusdf2.loc[cusdf2['age_class']=='40-50대','price'] = cusdf2.loc[cusdf2['age_class']=='40-50대','price'].apply(spend)
# 계절별 소비증가 및 감소 함수 적용하기
cusdf2.loc[cusdf2['season']=='spring','price'] = cusdf2.loc[cusdf2['season']=='spring','price'].apply(spend)
cusdf2.loc[cusdf2['season']=='fall','price'] = cusdf2.loc[cusdf2['season']=='fall','price'].apply(spend)
cusdf2.loc[cusdf2['season']=='winter','price'] = cusdf2.loc[cusdf2['season']=='winter','price'].apply(dspend)
cusdf2.loc[cusdf2['season']=='summer','price'] = cusdf2.loc[cusdf2['season']=='summer','price'].apply(dspend)
```

```python
# amount 열 추가하기 amount = price * quantity
cusdf2['amount']=cusdf2['price']*cusdf2['quantity']
## cusdf2['amount'].apply(lambda x:'%.2f'% x).inplace=True
cusdf2.head(20)
```

```python
# 고객번호,성별, 지역 날짜 나이별 요일별 어마운트 시즌만 남기고 분서해보기
cusdf3=cusdf2.copy()
cusdf3=cusdf3[['cusno','gender','age_class','region','amount','date','day','season','month']]
cusdf3
```

```python
# pivot 으로 데이터 정렬하기
pv_t = pd.pivot_table(cusdf3,
                       index=['gender','age_class'],
                       columns = 'season',
                       values = 'amount',
                       aggfunc = 'mean')
pv_t
```
