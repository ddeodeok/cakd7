### 범주화

```python
#범주화
def get_category(x):
    cat = ''
    if x<= 10:
        cat = 'Child'#10살이하
    elif x<= 30:
        cat = 'Young adult'#11-30살
    elif x<= 60:
        cat = 'Adult'#30-60살
    else:
        x =  'Elderyl'#60살이상
    return cat

df['age_class'] = df.age.apply(lambda x:get_category(x))
df[['age','age_class']].head()
    
```

```python
def upper_(x):
    return x.upper()
df['sex'] = df.sex.apply(lambda x :upper_(x))
df[['sex']].head()
```

```python
cars = pd.read_csv('mtcars.csv',encoding='cp949')
cars.head()
```

```python
cars_qsec = cars[['qsec']]
qsec_max = cars_qsec.max()
qsec_min = cars_qsec.min()
qsec_mid = (qsec_max + qsec_min)/2  #중간값
counters = cars_qsec[cars_qsec > qsec_mid]  # nan값나옴
overnum = counters.dropna()  # nan값 지우는 함수 (.dropna())
len(overnum)
```
