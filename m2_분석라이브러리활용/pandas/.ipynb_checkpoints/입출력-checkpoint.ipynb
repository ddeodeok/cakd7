```python
# csv
import numpy as np
import pandas as pd
ar1 = np.arange(10).reshape(2,5)
df = pd.DataFrame(ar1,columns=list('abcde'))
df.to_csv('test.csv')
```

```python
 # index_col=0 은 인덱스 상 언네임 뜨는걸 없애는것
df = pd.read_csv('test.csv',index_col=0)
df
```

```python
# json (딕셔너리 파일로 되있음)
df1 = df.copy()
df1.to_json('test1.json')
df1_json = pd.read_json('test1.json')
df1_json
```

```python
# 바이너리 파일 읽고 쓰기
df2 = df.copy()
df2.to_pickle('test2.pkl')
df2_pk1 = pd.read_pickle('test2.pkl')
df2_pk1
```
