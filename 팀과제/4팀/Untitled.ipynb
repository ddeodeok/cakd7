### 중부원점(EPSG:2097) 위도 경도(WSG84)로 변환

```python
!pip install pyproj
```

```python
from pyproj import Proj, transform
```

```python
# Projection 정의
# 중부원점(Bessel): 서울 등 중부지역 EPSG:2097
proj_1 = Proj(init='epsg:2097')

# WSG84 경위도: GPS가 사용하는 좌표계 EPSG:4326
proj_2 = Proj(init='epsg:4326')

converted = transform(proj_1, proj_2, df1['좌표정보(X)'].values, df1['좌표정보(Y)'].values)
df1['lon'] = converted[0]
df1['lat'] = converted[1]
```

```python
df1['나라']='대한민국'
df1['도시']='서울특별시'
df1.head()
```
